```python
# ======================
# 2. Import library
# ======================
from ultralytics import YOLO
from roboflow import Roboflow
import pandas as pd
```

```python
# ======================
# 3. Download dataset dari Roboflow
# ======================
# Ganti "YOUR_API_KEY" dengan API key Roboflow kamu
rf = Roboflow(api_key="eXH7AqWgpGvtwtKyYQx3")
project = rf.workspace("bisindo-compvis").project("bisindo-ixjpi-zpqau")
version = project.version(3)
dataset = version.download("yolov8")

# Lokasi yaml dataset
data_yaml = dataset.location + "/data.yaml"
print("Dataset config:", data_yaml)
```

```python
# ======================
# 4. Definisikan model yang mau dibandingkan
# ======================
model_names = ["yolov8m.pt", "yolov11m.pt"]
results = []
```

```python
# ======================
# 5. Training & Evaluasi tiap model
# ======================
for name in model_names:
    print(f"\nTraining {name} ...\n")
    model = YOLO(name)

    # Train
    model.train(
        data=data_yaml,
        epochs=25,
        imgsz=640
    )

    # Evaluate
    metrics = model.val(data=data_yaml)

    # Simpan hasil
    results.append({
        "Model": name,
        "mAP50": metrics.box.map50,
        "mAP50-95": metrics.box.map,
        "mAP75": metrics.box.map75,
        "Per-class mAP": metrics.box.maps
    })
```

```python
print("\n=== Training YOLOv11m ===\n")
model = YOLO("yolo11m.pt")

model.train(
  data=data_yaml, # path ke data.yaml
  epochs=25, # jumlah epoch (ubah sesuai kebutuhan)
  imgsz=640, # ukuran gambar
  batch=16, # ukuran batch (sesuaikan dg GPU)
  name="bisindo_yolov11m", # nama folder output
  resume=False # pastikan tidak lanjut dari checkpoint lama
)
```

```python
print("\n=== Evaluasi YOLOv11m ===\n")
metrics = model.val(data=data_yaml)

# Simpan hasil ke list dictionary

results = [{
  "Model": "yolo11m.pt",
  "mAP50": metrics.box.map50,
  "mAP50-95": metrics.box.map,
  "mAP75": metrics.box.map75,
  "Per-class mAP": metrics.box.maps
}]

# ======================
# 5. Bandingkan hasil model (tampilan seperti tabel)
# ======================

import pandas as pd

df = pd.DataFrame(results)
print("\n📊 Perbandingan Model:\n", df)
```
